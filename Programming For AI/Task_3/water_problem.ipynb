```python
x = 0
y = 0
goal = 5

x_capacity = int(input("Enter capacity of jug x: "))
y_capacity = int(input("Enter capacity of jug y: "))

def menu():
    print("1. Fill x")
    print("2. Fill y")
    print("3. Empty x")
    print("4. Empty y")
    print("5. Pour x into y until x is empty")
    print("6. Pour x into y until y is full")
    print("7. Pour y into x until y is empty")
    print("8. Pour y into x until y is empty")
    
while True:
    menu()
    choose = int(input("Choose the operation"))
    if choose == 1:
        x = x_capacity
        y = y
    elif choose == 2:
        y = y_capacity 
        x = x
    elif choose == 3:
        x = 0
        y = y
    elif choose == 4: 
        y = 0
        x = x 
    elif choose == 5:   
        if x + y > y_capacity:
            y = y_capacity
            x = 0
        else:
            y = y + x
            x = 0
    elif choose == 6:
        if x + y > y_capacity:
            x = x - (y_capacity - y)
            y = y_capacity
        else:
            y = y + x
            x = 0
    elif choose == 7:
        if x + y > x_capacity:
            x = x_capacity
            y = 0
        else:
            x = x + y
            y = 0
    elif choose == 8:
        if x + y > x_capacity:
            y = y - (x_capacity - x)
            x = x_capacity
        else:
            x = x + y
            y = 0
    else:
        print("Invalid choice")
        continue
    print(f"x: {x}, y: {y}")
    if x == goal or y == goal:
        break
```
